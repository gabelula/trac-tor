```python
# All tickets for Sponsor 19
csv_path='/Users/gaba/Code/torproject/trac/datasets/S2_tickets.csv'

import pandas as pd
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np

csv_file = pd.read_csv(csv_path)
```
